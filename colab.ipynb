```python
# ====================================================================
# SEL 3: PROSES TRAINING MODEL YOLOv8 SEGMENTATION
# ====================================================================
from ultralytics import YOLO

# 1. Memuat model prapelatihan YOLOv8 Nano Segmentation (versi paling ringan & cepat)
model = YOLO('yolov8n-seg.pt')

# 2. Memulai proses Fine-Tuning / Training
# Variabel 'data=path_dataset_yaml' mengambil otomatis lokasi dari Sel 2
model.train(
    data=path_dataset_yaml,   # Membaca file data.yaml hasil download Roboflow secara otomatis
    epochs=20,                 # 20 Epoch sangat pas untuk kejar tayang UAS (estimasi 15-20 menit selesai)
    imgsz=640,                 # Standar resolusi pemrosesan gambar YOLO
    device=0                   # Menginstruksikan Colab memakai kartu grafis (GPU T4) agar super cepat
)

print("\n====================================================")
print("🏆 PROSES TRAINING SELESAI DENGAN SUKSES!")
print("====================================================")
print("Langkah selanjutnya:")
print("1. Lihat panel menu folder di sebelah kiri Google Colab.")
print("2. Masuk ke folder: runs -> segment -> train -> weights")
print("3. Klik kanan pada file 'best.pt' lalu pilih DOWNLOAD.")
print("4. Pindahkan file 'best.pt' tersebut ke folder proyek web di laptop lokal Anda!")
print("====================================================")
```
